```python
import matplotlib.pyplot as plt

from tbsfusion import TBSFusion

# Create interface
fusion = TBSFusion('/dev/ttyUSB0', default_address=1,
                   baudrate=115200, discard_echo=True)

# Get the RSSI from 5600 to 5900 MHz in 3 MHz steps
freqs, rssi = fusion.rssi_scan_range(5600, 5900, 3)

# Plot the RSSI
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(freqs, rssi)

ax.set_xlabel('Frequency (MHz)')
ax.set_ylabel('RSSI')
ax.set_title('TBS Fusion RSSI')
ax.set_ylim(0, 1.1)
ax.grid()
```
